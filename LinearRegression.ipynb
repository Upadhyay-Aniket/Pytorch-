```python
# Define model (inut, output size, forward pass)
# Construct loss and optimizer
# Training Loop
# - forward pass: compute prediction and loss
# - backward paa: gradients
# - update weights
# Install missing modules
!pip install sklearn

# Import missing modules
import numpy as np
from sklearn import datasets

# Fix the code
import torch
import torch.nn as nn
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt

# prepare data
# Generating a regression datasets
X_numpy, Y_numpy = datasets.make_regression(n_samples = 100, n_features = 1, noise=20, random_state= 1)

X = torch.from_numpy(X_numpy.astype(np.float32))
Y = torch.from_numpy(Y_numpy.astype(np.float32))
Y = Y.view(Y.shape[0], 1) # To reshape

n_samples, n_features = X.shape

# model
input_size = n_features
output_size = 1
model = nn.Linear(input_size, output_size)

# loss and optimizers
learning_rate = 0.01
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)

# trainig lopp
num_epochs =100
for epoch in range(num_epochs):
  #  forward pass and loss
  y_predicted = model(X)
  loss = criterion(y_predicted, Y)

  # backward pass
  loss.backward()

  # update
  optimizer.step()

  optimizer.zero_grad()

  if (epoch + 1) % 10 == 0:
    print(f'epoch: {epoch+1}, loss = {loss.item():.4f}')

# plot
predicted = model(X).detach().numpy()
plt.plot(X_numpy, Y_numpy, 'ro')
plt.plot(X_numpy, predicted, 'b')
plt.show()

# # Install missing modules
# !pip install sklearn

# # Import necessary modules
# import numpy as np
# import torch
# import torch.nn as nn
# from sklearn import datasets
# import matplotlib.pyplot as plt

# # Generate regression dataset
# X_numpy, Y_numpy = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=1)

# X = torch.from_numpy(X_numpy.astype(np.float32))
# Y = torch.from_numpy(Y_numpy.astype(np.float32))
# Y = Y.view(Y.shape[0], 1)  # Reshape Y

# n_samples, n_features = X.shape

# # Define the model
# model = nn.Linear(n_features, 1)

# # Loss function and optimizer
# learning_rate = 0.01
# criterion = nn.MSELoss()
# optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# # Training loop
# num_epochs = 100
# for epoch in range(num_epochs):
#     # Forward pass and loss calculation
#     y_predicted = model(X)
#     loss = criterion(y_predicted, Y)

#     # Backward pass and optimization
#     optimizer.zero_grad()
#     loss.backward()
#     optimizer.step()

#     if (epoch + 1) % 10 == 0:
#         print(f'Epoch: {epoch+1}, Loss: {loss.item():.4f}')

# # Convert tensors to NumPy arrays for plotting
# X_numpy = X.numpy()
# Y_numpy = Y.numpy()
# predicted = model(X).detach().numpy()

# # Plot
# plt.scatter(X_numpy, Y_numpy, color='red', label='Original data')
# plt.plot(X_numpy, predicted, color='blue', label='Predicted data')
# plt.xlabel('X')
# plt.ylabel('Y')
# plt.title('Linear Regression')
# plt.legend()
# plt.show()
```
